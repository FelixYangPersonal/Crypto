```
from urllib.request import Request, urlopen
from datetime import datetime
from IPython.display import HTML
import time
import json
import pandas as pd
import base64

url_id = '填入ID的URL'
r_id = Request(url_id, headers={'User-Agent': 'Mozilla/5.0'})
data_id = json.loads(urlopen(r_id).read())

list_id = {}
columns_id = []
info_id = []

for j in range(len(data_id['data']['cryptoCurrencyMap'])):
  values = []
  for k, v in data_id['data']['cryptoCurrencyMap'][j].items():
    if j == 1:
      columns_id.append(k)
    values.append(v)
    
    if k == 'symbol':
      list_id.setdefault(v, )      
      list_id[v] = str(values[0])
      break
  info_id.append(values)

df_id = pd.DataFrame.from_dict(info_id)
df_id.columns = columns_id
print(df_id)

time_start = str(int(time.mktime(datetime(2013, 8, 4).timetuple())))
time_end = str(int(time.mktime(datetime(2021, 1, 13).timetuple())))

symbol = '填入虛幣symbol'
url_history = f'填入history的URL'
r_history = Request(url_history, headers={'User-Agent': 'Mozilla/5.0'})
data_history = json.loads(urlopen(r_history).read())

columns_history = []
info_history = []

def download_csv(df, title='Download CSV file', filename='data.csv'):
  csv = df.to_csv(index=False)
  payload = base64.b64encode(csv.encode()).decode()
  html = '<a download="{filename}" href="data:text/csv;base64,{payload}" target="_blank">{title}</a>'
  html = html.format(payload=payload,title=title,filename=filename)
  return HTML(html)

for j in range(len(data_history['data']['quotes'])):
  values = []
  for k, v in data_history['data']['quotes'][j]['quote']['USD'].items():
    if j == 1:
      columns_history.append(k)
    if k == 'timestamp':
      v = datetime.strptime(v, '%Y-%m-%dT%H:%M:%S.%fz').strftime('%Y-%m-%d')
    values.append(v)
  info_history.append(values)

df_history = pd.DataFrame.from_dict(info_history)
df_history.columns = columns_history
print(df_history)
download_csv(df_history, title=f'Download {symbol} CSV file', filename=f'{symbol}.csv')
```
